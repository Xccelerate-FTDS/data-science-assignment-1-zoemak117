```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('spotify_youtube_sample.csv', index_col=[0])
data

data.isna().sum()
# Album_type
# Single: One song
# Album: Multiple songs from the same singer, offering a comprehensive experience
# Compilation: Several separate recordings by either one performer or several performers

# Spotify Uri
# A URI is a code you can use for embedding, and searching within the Spotify app.

# Danceability
# Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity.

# Energy
# Energy in music is a perceptible measure of intensity and activity.

# Key
# In music a key is the major or minor scale around which a piece of music revolves. There are 12 keys in total.

# Loudness
# The perceived “strength” of a sound on a scale from quiet to loud.

# Speechiness
# Detects the presence of spoken words in a track

# Acousticness
# A confidence measure from 0.0 to 1.0 of whether the track has no amplification of sound except for one microphone

# Instrumentalness 
# Predicts whether a track contains vocals. Instrumentalness is measured on a scale of 0 (likely contains vocal content) to 100 (likely contains no vocal content)

# Liveness
# Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live.

# Valence
# musical positivity conveyed by a piece of music

# Tempo
# speed or pacing of a piece of music

# Duration_ms
# The duration of the track in milliseconds
```

```python
data = data.dropna()
data.shape
data.columns
data.head()
```

```python
data.describe()
```

```python
#Top 5 most popular songs by streaming
top_five_songs = data.sort_values(by="Stream",ascending=False)[["Artist", "Track", "Stream"]][:5]
top_five_songs

top_five_songs.plot.barh(x = "Track",y = "Stream")
plt.xlabel("Stream")
plt.ylabel("Track")
plt.show()
```

```python
#Top 5 most popular songs by YouTube Views
top_five_songs_views = data.sort_values(by="Views",ascending=False)[["Artist", "Track", "Views"]][:5]
top_five_songs_views

top_five_songs_views.plot.barh(x = "Track",y = "Views")
plt.xlabel("Views")
plt.ylabel("Track")
plt.show()
```

```python
#Top 5 artists
top_five_artists = data.groupby("Artist").sum().sort_values(by="Stream",ascending=False)[["Track", "Stream","Views"]][:5]
top_five_artists
```

```python
#Top 5 most danceable songs
top_dance_songs = data.sort_values(by="Danceability",ascending=False)[["Artist", "Track", "Stream","Views"]][:5]
top_dance_songs
```

```python
#Artists with the most danceable songs
top_artist_dance_songs = data.groupby("Artist").mean("Danceability").sort_values(by="Danceability",ascending=False)[["Danceability","Stream","Views"]][:5]
top_artist_dance_songs
```

```python
streamview = data[["Views", "Stream"]].corr()
streamview

#not very correlated
```

```python
#Stream Correlation Matrix
cm_stream = data[["Danceability", "Energy", "Loudness", "Speechiness", "Acousticness", "Instrumentalness", "Liveness", "Valence", "Tempo", "Duration_ms", "Stream"]].corr()

# Creating heatmap to depict correlation
plt.figure(figsize = (27, 27))
sns.heatmap(cm_stream, annot = True, cmap = 'coolwarm', linewidth = 0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
#Views Correlation Matrix
cm_view = data[["Danceability", "Energy", "Loudness", "Speechiness", "Acousticness", "Instrumentalness", "Liveness", "Valence", "Tempo", "Duration_ms", "Views"]].corr()

# Creating heatmap to depict correlation
plt.figure(figsize = (27, 27))
sns.heatmap(cm_view, annot = True, cmap = 'coolwarm', linewidth = 0.5)
plt.title('Correlation Matrix')
plt.show()
```
